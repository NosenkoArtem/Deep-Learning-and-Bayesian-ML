```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.path as path
import matplotlib.mlab as mlab
import scipy.stats as stats
```

### Случайное блуждание под графиком функции

```python
delta = 0.025
X, Y = np.meshgrid(np.arange(-5.5, 2.0, delta), np.arange(-4.5, 3.5, delta))

# формируем исходное распределение
z1 = stats.multivariate_normal([0,0],[[1.0,0],[0,1.0]])
z2 = stats.multivariate_normal([-2,-2],[[1.5,0],[0,0.5]])
def z(x): return 0.4*z1.pdf(x) + 0.6*z2.pdf(x)

Z1 = mlab.bivariate_normal(X, Y, 1.0, 1.0, 0.0, 0.0)
Z2 = mlab.bivariate_normal(X, Y, 1.5, 0.5, -2, -2)
Z = 0.4*Z1 + 0.6*Z2

Q = stats.multivariate_normal([0,0],[[0.02,0],[0,0.02]])
r = [-2, 2]
samples = [r]
for i in range(0, 5000):
#     выбираем случайное число из нормального распределения
	rq = Q.rvs() + r
	a = z(rq)/z(r)
#   семплируем случайное число из биномиального распределения, пропорционально a
	if np.random.binomial(1, min(a, 1), 1)[0] == 1:
		r = rq
		samples.append(r)
```

```python
codes = np.ones(len(samples), int) * path.Path.LINETO
codes[0] = path.Path.MOVETO
p = path.Path(samples,codes)

fig, ax = plt.subplots()
ax.contour(X, Y, Z)
ax.add_patch(patches.PathPatch(p, facecolor='none', lw=0.5, edgecolor='red'))
plt.grid()
```
